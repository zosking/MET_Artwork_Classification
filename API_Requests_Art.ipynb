```python
import requests
import json
import numpy as np
import pandas as pd
from pandas.io.json import json_normalize
```

# Downloading Data for Each Work of Art

```python
#generate list of unique object IDs used to identify each work of art
url = 'https://collectionapi.metmuseum.org/public/collection/v1/objects'

objects = requests.get(url, headers={ "Accept" : "application/json"})

objects_json = objects.json()
object_IDs = objects_json['objectIDs']
```

```python
#request all information for each work of art in the Met public database
url = 'https://collectionapi.metmuseum.org/public/collection/v1/objects/'

y = 0
all_art = []

for piece in range(len(object_IDs)):
    #since each piece needs to be requested individually, takes the 
    obj_url = url + str(object_IDs[y])
    object_info = requests.get(obj_url, headers={ "Accept" : "application/json"})
    artwork = object_info.json()
    all_art.append(artwork)
    y += 1
    #print(str(y) + ' ' + ' loaded')
    
all_art_df = pd.DataFrame(all_art)
all_art_df.to_csv('art_df.csv')
```

```python
art_df = pd.read_csv('art_df.csv', index_col='objectID',lineterminator='\n', low_memory=False)
```

```python
df = art_df.replace('',np.NaN)
df['objectID'] = df['Unnamed: 0']
df.set_index('objectID', inplace=True)
df = df[df.primaryImageSmall.notna()]
```

# Downloading Images of Art Work

```python
#generate list of hyperlinks to images of the art work
links = df.primaryImageSmall
```

```python
#download pictures locally
import urllib.request

x = 0
bad_links = 0

for link in links.index:
    try:
        urllib.request.urlretrieve(links[int(link)], filename= 'Images/'+ str(int(link)) + '_' + links[int(link)].split('/')[-1])
        x += 1
    except:
        bad_links += 1
        pass
```
